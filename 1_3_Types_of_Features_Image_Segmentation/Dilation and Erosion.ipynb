```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

### Dilation

```python
# Reads in a binary image
image = cv2.imread('images/j.png', 0)

# Create a 5x5 kernel of ones
kernel = np.ones((5,5), np.uint8)

# Dilate the image
dilation = cv2.dilate(image, kernel, iterations = 1)

plt.imshow(dilation, cmap='gray')
```

### Erosion

```python
erosion = cv2.erode(image, kernel, iterations=1)

plt.imshow(erosion, cmap='gray')
```

### Visualizing Erosion and Dilation

```python
_, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))

ax1.set_title('Erosion')
ax1.imshow(erosion, cmap='gray')

ax2.set_title('Original')
ax2.imshow(image, cmap='gray')

ax3.set_title('Dilation')
ax3.imshow(dilation, cmap='gray')
```

### Opening

```python
op_image = cv2.imread('images/open.png')

opening = cv2.morphologyEx(op_image, cv2.MORPH_OPEN, kernel)

_, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title('Original')
ax1.imshow(op_image, cmap='gray')

ax2.set_title('Opening')
ax2.imshow(opening, cmap='gray')
```

### Closing

```python
cl_image = cv2.imread('images/close.png')

closing = cv2.morphologyEx(cl_image, cv2.MORPH_CLOSE, kernel)

_, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.set_title('Original')
ax1.imshow(cl_image, cmap='gray')

ax2.set_title('Closing')
ax2.imshow(closing, cmap='gray')
```
